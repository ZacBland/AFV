```python
import ee

# Initialize the Earth Engine module.
ee.Initialize()

# Your code to specify the image.
stillwaterAirport = ee.Geometry.Point([-97.086237, 36.162738])
dataset = ee.ImageCollection('LANDSAT/LC08/C01/T1') \
    .filterBounds(stillwaterAirport) \
    .filterDate('2023-01-01', '2023-12-31') \
    .sort('CLOUD_COVER') \
    .first()

# Define the visualization parameters.
rgbVisualization = {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 30000}

# Convert the image to an RGB visualization.
rgbImage = dataset.visualize(**rgbVisualization)

# Define the region to export.
region = stillwaterAirport.buffer(10000).bounds().getInfo()['coordinates']

# Start the export to Google Drive.
task = ee.batch.Export.image.toDrive(
    image=rgbImage,
    description='StillwaterAirport_RGB',
    scale=30,
    region=region,
    fileFormat='GeoTIFF'
)
task.start()

# Print the URL to the task (optional, for debug purposes).
print(task.status())
```

```python
import ee

# Initialize the Earth Engine module.
ee.Initialize()

# Define the polygon geometry.
area = ee.Geometry.Polygon(
    [[
        [-97.07643951347823, 36.15077249982036],
        [-97.07643951347823, 36.17363997553162],
        [-97.09497894218917, 36.17363997553162],
        [-97.09497894218917, 36.15077249982036]
    ]]
)

# Load the Sentinel-2 ImageCollection.
sentinel2 = ee.ImageCollection('COPERNICUS/S2')

# Filter the collection by the ROI and date.
filtered_sentinel2 = sentinel2.filterBounds(area).filterDate('2020-01-01', '2023-12-31').sort('CLOUD_COVER')

# Check if the collection has any images after filtering.
if filtered_sentinel2.size().getInfo() > 0:
    # Get the first image from the filtered collection.
    image_to_export = filtered_sentinel2.first()

    # Cast all bands to UInt16.
    image_to_export = image_to_export.toUint16()

    # Define the export parameters.
    export_params = {
        'image': image_to_export,
        'description': 'LandInsidePolygon',
        'scale': 30,  # The smallest scale for Sentinel-2 is 10 meters for RGB and NIR bands
        'region': area.getInfo()['coordinates'],  # Use the polygon coordinates.
        'fileFormat': 'GeoTIFF',
        'maxPixels': 1e13  # Specify the maxPixels to avoid memory issues.
    }

    # Create an export task.
    task = ee.batch.Export.image.toDrive(**export_params)

    # Start the export task.
    task.start()

    # Print the URL to the task (optional, for debug purposes).
    print(task.status())
else:
    print('No images found in the ImageCollection for the specified filters.')
```
